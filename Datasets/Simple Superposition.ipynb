```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10, 10))
for r in range(4):
    for c in range(4):
        ax = axes[r, c]
        plot_input_image(objects[4*r + c], ax)
        ax.set_xlabel('Object {}'.format(4*r + c + 1))
```

```python
all_pairs = np.zeros((1, len(objects)*(len(objects)-1)//2, 8, 8))
all_pair_groups = np.zeros_like(all_pairs)

k = 0
for i in range(len(objects)):
    for j in range(i + 1, len(objects)):
        all_pairs[0, k, :, :] = objects[i] + objects[j]
        all_pair_groups[0, k][objects[i] != 0] = 1
        all_pair_groups[0, k][objects[j] != 0] = 2
        all_pair_groups[0, k][all_pairs[0, k] > 1] = 0
        k += 1

all_pairs = (all_pairs != 0).astype(np.float32)
```

```python
# WARNING: SLOW! for some reason this plot takes a couple of minutes. Feel free to skip
fig, axes = plt.subplots(nrows=16, ncols=16, figsize=(20, 20))
k = 0
for r in range(16):
    for c in range(16):
        ax = axes[r, c]
        if c <= r:
            ax.set_visible(False)
            continue
        plot_groups(all_pair_groups[0, k], ax)
        k += 1
```

# Save as Dataset

```python
import h5py
import os
import os.path

data_dir = os.environ.get('BRAINSTORM_DATA_DIR', '.')
```

```python
with h5py.File(os.path.join(data_dir, 'simple_superpos.h5'), 'w') as f:
    single = f.create_group('train_single')
    single.create_dataset('default', data=np.array(objects).reshape(1, 16, 8, 8, 1), compression='gzip')
    single.create_dataset('groups', data=np.array(objects).reshape(1, 16, 8, 8, 1), compression='gzip')
    
    pairs = f.create_group('test')
    pairs.create_dataset('default', data=all_pairs.reshape(1, -1, 8, 8, 1), compression='gzip')
    pairs.create_dataset('groups', data=all_pair_groups.reshape(1, -1, 8, 8, 1), compression='gzip')
```
